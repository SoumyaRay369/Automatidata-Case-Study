#### From the above I have concluded that there are some null values in the 'label' column

#### **How many iPhone users had null values and how many Android users had null values?**

```python
df[df['label'].isna()]['device'].value_counts()
```

#### **Percentage of iPhone and Android users with null values**

```python
null_labels = df[df['label'].isna()]
null_labels
```

```python
null_labels['device'].value_counts(normalize=True).reset_index(name = 'percentage')
```

#### **Examine the count of users who were churned vs. retained**

#### **Median value of all the columns for churned and retained users**

#### **Calculate the median drive per km for each group**

```python
data = df.drop(columns=['device'])
data
```

```python
data['drive_per_km'] = data['driven_km_drives'] / data['drives']
# data.drop(columns=['driver_per_km'], inplace=True)
data
```

```python
data.groupby('label')['drive_per_km'].median()
```

#### **How many km per day does each category of driver drive?**

```python
data['km_per_day'] = data['driven_km_drives'] / data['driving_days']
data
data.groupby('label')['km_per_day'].median()
```
