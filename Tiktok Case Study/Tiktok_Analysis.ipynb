#### **What are the different values of each claim status and how many them are there in the data?**

```python
df['claim_status'].value_counts().reset_index(name='count')
```

#### **Average view count of videos with respect to their claim status**

```python
df.groupby('claim_status')['video_view_count'].mean().reset_index(name='average video view count')
```

#### **Count for each combination of claim_status and author_ban_status**

```python
df.groupby(['claim_status', 'author_ban_status']).size().reset_index(name='count')
```

#### **Mean video_share_count for each author_ban_status**

```python
df.groupby(['author_ban_status'])[['video_share_count', 'video_like_count', 'video_view_count']].agg(['mean', 'median', 'count'])
```

#### **Create likes_per_view, shares_per_view and comments_per_view columns**

```python
df['likes_per_view'] = df['video_like_count'] / df['video_view_count']
df['shares_per_view'] = df['video_share_count'] / df['video_view_count']
df['comments_per_view'] = df['video_comment_count'] / df['video_view_count']
df.head()
```

#### **Analyse the claim_status and author_ban_status on these newly created metrics**

```python
df.groupby(['claim_status', 'author_ban_status'])[['likes_per_view', 'shares_per_view', 'comments_per_view']].agg(['mean', 'median', 'count'])
```
